```python
arr1 = np.arange(10, 50)
print("original array:\n",arr1)
matrix1 = arr1.reshape(4, 10)
print("\nReshaped 4x10 matrix:\n",matrix1)
last_column = matrix1[:, -1]
print("\nLast column:\n",last_column)
```

```python
mat_fives = np.full((3, 3), 5)
mat_1to9 = np.arange(1, 10).reshape(3, 3)
print("Matrix of 5s:\n", mat_fives)
print("\nMatrix 1 to 9:\n", mat_1to9)

added_mat = mat_fives + mat_1to9
print("\nAdded Matrix:\n", added_mat)

multiplied_mat = mat_fives * mat_1to9
print("\nElement-wise Multiplied Matrix:\n", multiplied_mat)

transpose_mat = multiplied_mat.T
```

```python
import numpy as np
arr3 = np.random.randint(10, 101, 15)
print("Original Array:")
print(arr3)

count = np.sum(arr3>50)

arr3[arr3>50] = -1

print("\nModified Array:")
print(arr3)

print("\nNumber of values replaced:", count)
```

```python
matrix4 = np.random.rand(5, 4)
print("Original 5x4 Matrix:\n", matrix4)

row_sums = matrix4.sum(axis=1)
max_sum_row_index = np.argmax(row_sums)
print("\nSums of each row:", row_sums)
print("Row index with maximum sum:", max_sum_row_index)

row_min = matrix4.min(axis=1, keepdims=True)
row_max = matrix4.max(axis=1, keepdims=True)

normalized_matrix = (matrix4 - row_min) / (row_max - row_min)
print("\nNormalized Matrix:\n", normalized_matrix)
```

```python
import numpy as np
arr5 = np.array([1, 3, 5, 3, 7, 9, 1, 5, 5])
print("Original Array:", arr5)

unique_elements = np.unique(arr5)
print("Unique Elements:", unique_elements)

elements, counts = np.unique(arr5, return_counts=True)
duplicates = elements[counts > 1]
print("Duplicate Elements:", duplicates)
```

```python
matrix6 = np.arange(1, 37).reshape(6, 6)
print("Original 6x6 Matrix:\n", matrix6)

submatrix = matrix6[1:5, 1:5]
print("\nMiddle 4x4 Submatrix:\n", submatrix)

even_numbers = submatrix[submatrix % 2 == 0]
print("\nEven Numbers from Submatrix:\n", even_numbers)
```

```python
arr7 = np.arange(12).reshape(3, 4)
print("Original Matrix (3x4):\n", arr7)

reshaped_2x6 = arr7.reshape(2, 6)
print("\nReshaped Matrix (2x6):\n", reshaped_2x6)

flattened_arr = reshaped_2x6.flatten()
print("\nFlattened 1D Array:\n", flattened_arr)

final_matrix = flattened_arr.reshape(4, 3)
print("\nFinal Reshaped Matrix (4x3):\n", final_matrix)
```

```python
a = np.array([[1, 2], [3, 4]])
print("Initial Matrix:\n", a)

a_with_row = np.append(a, [[5, 6]], axis=0)
print("\nAfter Appending New Row:\n", a_with_row)

new_column = np.array([[7], [8], [9]])
final_a = np.append(a_with_row, new_column, axis=1)
print("\nAfter Appending New Column:\n", final_a)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.array([7, 8, 9])

joined_arr = np.concatenate((a, b, c))
print("Joined 1D Array:\n", joined_arr)

matrix9 = joined_arr.reshape(3, 3)
print("\nReshaped 3x3 Matrix:\n", matrix9)

diagonal_elements = np.diag(matrix9)
print("\nDiagonal Elements:\n", diagonal_elements)
```

```python
arr10 = np.arange(1, 21)
print("Original Array:\n", arr10)

filtered_values = arr10[(arr10 % 3 == 0) | (arr10 % 5 == 0)]
print("\nValues divisible by 3 or 5:\n", filtered_values)

count_values = len(filtered_values)
print("\nTotal count of these values:", count_values)
```
