Part C – Data Transformation

Task 6

Rename columns wherever necessary so that all datasets have a common column

name for merging.

Hint:

● Example: Country → country

```python
population.rename(columns={
    "Country": "country"
}, inplace=True)
population.head()
gdp.rename(columns={
    "Country": "country"
}, inplace=True)
gdp.head()
internet.rename(columns={
    "Country": "country"
}, inplace=True)
internet.head()
literacy.rename(columns={
    "Country": "country"
}, inplace=True)
literacy.head()
print(population.columns)
print(gdp.columns)
print(internet.columns)
print(literacy.columns)
```

Task 7 – Convert Numeric Columns

```python
population["population"] = pd.to_numeric(population["population"])
gdp["GDP"] = pd.to_numeric(gdp["GDP"])
internet["internet_users"] = pd.to_numeric(internet["internet_users"])
literacy["literacy_rate"] = pd.to_numeric(literacy["literacy_rate"])
```

Task 8 – Merge All Datasets

```python
data = population.merge(gdp,on="country")
data = data.merge(internet,on="country")
data = data.merge(literacy, left_on="country", right_on="name")
data.drop(columns=["name"], inplace=True)
data.head()
```

Task 9 – Calculated Columns

Internet Penetration

```python
data["internet_penetration"] = (
    data["internet_users"] /
    data["population"]
)*100
```

GDP Per Capita

```python
data["gdp_per_capita"] = (
    data["GDP"] /
    data["population"]
)
```

Round

```python
data["internet_penetration"] = data["internet_penetration"].round(2)
data["gdp_per_capita"] = data["gdp_per_capita"].round(2)
```

Part- D --> Load Data

Task 10 – Store in SQLite

```python
conn = sqlite3.connect("country.db")
data.to_sql(
    "country_statistics",
    conn,
    if_exists="replace",
    index=False
)
```

Task 11 – Verify Database

```python
pd.read_sql(
    "SELECT * FROM country_statistics LIMIT 5",
    conn
)
```

ETL Process Summary

1. Data Sources Used

Data was collected from four different file formats: CSV (Population), Excel (GDP), JSON (Internet Users), and XML (Literacy Rate).

Each dataset was explored to identify rows, columns, data types, missing values, and duplicate records. Column names were standardized, and numeric columns were converted to the appropriate data types.

3. Transformations Performed

The datasets were merged using the common country column. New fields such as Internet Penetration (%) and GDP per Capita were calculated for analysis.

4. Database Loading Process

The final merged dataset was stored in a SQLite database using Pandas to_sql() function. The processed dataset was also exported as a CSV file.

5. Key Insights Obtained

The analysis identified the countries with the highest GDP, largest population, highest GDP per capita, and highest internet penetration. It also calculated the average literacy rate and total population, providing useful insights into global development indicators.

6. Conclusion

The ETL process successfully extracted data from multiple sources, transformed it into a consistent format, and loaded it into a database for efficient analysis and reporting.
