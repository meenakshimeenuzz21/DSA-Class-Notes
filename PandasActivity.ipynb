```python
data['Age'] = data['Age'].fillna(data['Age'].median())
data
```

```python
def age_group(Age):
    if Age < 18:
      return "Child"
    elif Age <=60:
      return "Adult"
    else:
      return "Senior"
data['age_group'] = data['Age'].apply(age_group)
data[['Age', 'age_group']].head()
```
